### Abbildung 1

```python
from pynbTurtle import Turtle

def part_circle(x):
    for i in range(int(180/x)):
        t.forward(2)
        t.right(1)

def flower(leafs,circle):
    for j in range(leafs):
        for i in range(2):
            part_circle(circle)
            t.right(180-int(180/4))
        t.right(int(360/leafs))

t = Turtle()
t.speed(10)
flower(20,4)
```

```python
from pynbTurtle import Turtle 
 	 	 	 	 
def zacke():
    t.forward(100) 
    t.left(120) 	 	 
    t.forward(100) 
    t.right(60) 	 	 
 	 	 	 
def zacke_rot():
    t.pencolor("red")
    zacke() 	 	 
	 	 	 	 
def zacke_blau(): 
    t.pencolor("blue") 
    zacke() 	 	 
 	 	 	 
def stern():
    for i in range(3):
        zacke_rot()
        zacke_blau()
 
t=Turtle()
t.speed(10)	
t.showcode() 
stern()
t.penup()
t.left(120)
t.forward(100)
t.right(120)
```

### Abbildung 4

```python
from pynbTurtle import Turtle

def linie():
    t.forward(40)
    t.backward(40)
    
def spitze():
    linie()
    t.left(45)
    linie()
    t.right(90)
    linie()
    t.left(45)
    
def ast():
    t.forward(40)
    spitze()
    t.backward(40)
    
def flocke():
    for i in range(6):
        ast()
        t.right(60)

t=Turtle()
t.speed(10)
t.showcode()
flocke()
```

### Beispiel 1

```python
from pynbTurtle import Turtle

t = Turtle()
t.speed(10)  # Setze die Geschwindigkeit auf 10 (schnellste)
t.forward(100) # Bewege die Turtle 100 Einheiten vorwärts
t.left(90)      # Drehe die Turtle um 90 Grad nach links
t.backward(50)  # Bewege die Turtle 50 Einheiten rückwärts
t.right(60)     # Drehe die Turtle um 60 Grad nach rechts
t.penup()   # Hebe den Stift an, um keine Linie zu zeichnen
t.forward(200)  # Bewege die Turtle 200 Einheiten vorwärts
t.pendown() # Setze den Stift ab, um eine Linie zu zeichnen
t.pencolor("yellow") # Setze die Stiftfarbe auf rot
t.backward(100) # Bewege die Turtle 100 Einheiten rückwärts
t.left(-90) # Drehe die Turtle um -90 Grad nach links -> Rechtsdrehung
t.pencolor("green") # Setze die Stiftfarbe auf rot
t.backward(50) # Bewege die Turtle 50 Einheiten rückwärts
t.setposition(200,200) # Verschiebe die Turtle an die Position (200,200)
```

### Beispiel 2

```python
from pynbTurtle import Turtle

def circle(steps): # Hier wird der Befehl für einen Kreis festgelegt
    for i in range(360):
        t.forward(steps) # Bewege die Turtle um steps Einheiten vorwärts
        t.right(1) # Drehe die Turtle um 90 Grad nach rechts
        
t = Turtle()
t.speed(10)
t.showcode('play') # showcode wird im Modus 'play' aufgerufen
circle(2)  # Zeichne einen Kreis mit Schrittgrösse 2
t.left(90) # Drehe die Turtle um 90 Grad nach links
t.forward(100) # Bewege die Turtle 100 Einheiten vorwärts
t.left(90)  # Drehe die Turtle um 90 Grad nach links
circle(1) # Zeichne einen Kreis mit Schrittgrösse 1
```

### Beispiel 3

```python
from pynbTurtle import Turtle

def quadrat(seite, farbe):
    t.pencolor(farbe)
    for i in range(4):
       t.forward(seite)
       t.right(90) 

def fenster(seite, farbe):
    for i in range(4):
       quadrat(seite, farbe)
       t.right(90) 

def mosaik(seite, farbe1, farbe2):
    fenster(2*seite, farbe1)
    t.forward(seite)
    t.left(90)
    t.forward(seite)
    fenster(seite, farbe2)
    t.backward(seite)
    t.right(90)
    t.backward(2*seite)
    t.right(90)
    t.forward(seite)
    fenster(seite, farbe2)

t=Turtle()
t.speed(10)
t.showcode()
mosaik(80, "blue", "red")
```
